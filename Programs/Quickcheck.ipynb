```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://mywebsite.com" # The URL you would like to check
```

```python
response = requests.get(url) # Asking the website for information

site_info = BeautifulSoup(response.content, "html.parser") # Parsing that website for information (if not HTML, may need to change the parser)

print(site_info.text) # Displaying the message from the website
```
